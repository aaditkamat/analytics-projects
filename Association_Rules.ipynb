```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
retail_df = pd.read_csv('./Online Retail.csv')
retail_df.head()
```

```python
# Processing the entire dataset is time consuming, hence we should consider
# only a certain number of transactions
no_invoices = -1

grouped_count = retail_df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().reset_index()
# grouped_count = grouped_count.loc[grouped_count['InvoiceNo'].isin(grouped_count['InvoiceNo'].unique()[: no_invoices]), :]

items_df = pd.pivot_table(grouped_count, values='Quantity', columns='Description', index='InvoiceNo', aggfunc='sum', fill_value=0)
items_df = items_df.applymap(lambda x: 0 if x == 0 else 1)
```

```python
# Since the probability of an item or groups of items being purchased is very low,
# because there are different kinds of items across transactions, we set the threshold
# for support as a low number
prob_df = apriori(items_df, min_support = 0.03, use_colnames = True)
prob_df
```

```python
 # We want to consider itemsets where lift is greater than 1
rules_df = association_rules(prob_df, metric='lift', min_threshold=1)
rules_df
```

The associations at the top make sense because you often buy cake cases and paper chain kits together to celebrate events like birthdays or weddings.
