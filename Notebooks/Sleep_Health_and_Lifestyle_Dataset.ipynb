# Does an increased physical activity level result in a better quality of sleep?

```python
import statsmodels.api as sm
```

```python
x = SHL_df[['Physical Activity Level', 'Daily Steps']]
y = SHL_df['Quality of Sleep']
X = sm.add_constant(x)
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

```python
SHL_df['Predicted Quality of Sleep'] = model.predict(results.params)
```
